```python
import numpy as np
from sklearn.naive_bayes import BernoulliNB
```

```python
q = np.genfromtxt('first_questions.csv', delimiter=',', skip_header=1, dtype='int')
s = np.genfromtxt('first_student.csv', delimiter=',', skip_header=1)
```

```python
b = np.max(q[:, 1])
a = np.zeros((s.shape[0], b))
m = np.zeros(b)
for i, j in enumerate(q):
    a[:, j[1] - 1] += s[:,i]
    m[j[1] - 1] += j[0]
percent = np.divide(a, m, out=np.zeros_like(a), where=m != 0)
percent
```
